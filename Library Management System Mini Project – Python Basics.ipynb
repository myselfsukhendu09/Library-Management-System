```python
# Step 1: Basic Structure:-

# Print a welcome message to the user when the program starts.
print("Welcome to the Library Management System")
```

```python
# Library structure

# Define a dictionary named 'library'.
# This dictionary will store information about books in the library.
library = {
 # Add a book entry with its title as the key
     "Python Basics": {
     "author": "Test",         # Store the author's name for the book
     "available": True,        # A boolean indicating whether the book is available for borrowing.
     "issued_to": None         # This stores the name of the person who has the book issued, Here None means no one has it. 
    },
# Add another book entry with a different title.
     "Data Science": {
     "author": "Test1",
     "available": False,        # This book is not available when someone has already borrowed it.
     "issued_to": "John"        # The name of the person who borrowed the book.
    }
}
# Define a function named 'display_books' to display all books with details inshow all books in the library.
def display_books():
    print("\n--- List of Books in Library ---")
    # Loop through each item (book) in the 'library' dictionary
    for title, details in library.items():     # 'title' is the name of the book (key), and 'details' is the inner dictionary (value).
        status = "Available" if details["available"] else f"Issued to {details['issued_to']}"
        print(f"Title: {title}, Author: {details['author']}, Status: {status}")
```

```python
# Step 2: Adding Books

# Define a function called 'add_book'
# This function allows the user to add a new book to the library
def add_book():
    title = input("Enter book title: ")
    author = input("Enter author name: ")
    if title in library:
        print("Book already exists in the library.")     # Check if the book title already exists in the 'library' dictionary.
    else:
        library[title] = {
            "author": author,     # Set the author.
            "available": True,    # Book is available by default.
            "issued_to": None     # Not issued to anyone yet.
        }
        print(f"Book '{title}' by {author} added successfully.")
```

```python
# Step 3: Issuing Books:-

# Define a function called 'issue_book'
# This function allows a user to issue a book from the library
def issue_book():
    title = input("Enter the book title to issue: ")
    if title in library:
        if library[title]["available"]:                                         # Check if the book is currently available for issuing.                    
            person = input("Enter the name of the person issuing the book: ")   # If available, ask for the name of the person who wants to issue it.
            library[title]["available"] = False                                 # Update the book's availability status to False if issued.
            library[title]["issued_to"] = person                                # Store the name of the person the book is issued to.
            print(f"Book '{title}' has been issued to {person}.")               # Confirm to the user that the book has been issued successfully.    
        else:
            print(f"Book '{title}' is already issued to {library[title]['issued_to']}.")
    else:
        print("Book not found in the library.")
```

```python
# Step 4: Returning Books:-

# Define a function called 'return_book'
# This function allows a user to return a previously issued book to the library
def return_book():
    title = input("Enter the book title to return: ")
    if title in library:
        if not library[title]["available"]:
                library[title]["available"] = True        # Set the book's status to available again
                person = library[title]["issued_to"]      # Get the name of the person who had issued the book 
                library[title]["issued_to"] = None        # Reset the 'issued_to' field to None since it's now returned
                print(f"Book '{title}' has been returned by {person}.")
        else:
            print(f"Book '{title}' was not issued.")
    else:
        print("Book not found in the library.")
```

```python
# Step 5: Login System:-

# Create a dictionary called 'credentials'
# This dictionary stores usernames as keys and their corresponding passwords as values
credentials = {
    "admin": "admin123",
    "librarian": "lib123"
}
# Define a function named 'login' to handle user login attempts
def login():
    attempts = 0
    while attempts < 3:
        #Ask the user to input their username
        username = input("Enter username: ")
        password = input("Enter password: ")
# use 'and' opareter whether the entered password matches the stored password.
        if username in credentials and credentials[username] == password:
            print(f"Login successful. Welcome, {username}!")
            return True
        else:
            attempts += 1
            print(f"Invalid credentials. Attempts left: {3 - attempts}")
    print("Too many failed attempts. Exiting.")
    return False
```

```python
# Step 6: View Issued Books:-

# This function will display a list of all books that are currently issued.
def view_issued_books():
    print("\n--- Issued Books ---")
    found = False
# 'title' is the book name, and 'details' is the dictionary of its info.
    for title, details in library.items():
        if not details["available"]:
            print(f"Title: {title}, Issued to: {details['issued_to']}")
            found = True
    if not found:
        print("No books are currently issued.")
```

```python
# Step 7: Main Menu:-
# define a function called login()
if login():
    while True:
        print("\n--- Library Management System Menu ---")
        print("1. View Books")           # Option to list all books.
        print("2. Add a Book")           # Option to add a new book.
        print("3. Issue a Book")         # Option to issue a book to someone.
        print("4. Return a Book")        # Option to return a previously issued book.
        print("5. View Issued Books")    # Option to list books that are currently issued.
        print("6. Exit")                 # Option to exit the system.
        choice = input("Enter your choice (1-6): ")
        if choice == "1":
            display_books()
        elif choice == "2":
            add_book()
        elif choice == "3":
            issue_book()
        elif choice == "4":
            return_book()
        elif choice == "5":
            view_issued_books()
        elif choice == "6":
            print("Exiting Library Management System. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 6.")          
```
